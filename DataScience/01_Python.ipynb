## 4. 📋 ListasLas listas son colecciones ordenadas que pueden contener múltiples elementos. ¡Son como estantes donde puedes guardar diferentes objetos! 📚### 🎯 **Características de las Listas:**- **📊 Ordenadas**: Mantienen el orden de los elementos- **🔄 Mutables**: Puedes modificarlas después de crearlas- **🎭 Versátiles**: Pueden contener diferentes tipos de datos- **📍 Indexadas**: Cada elemento tiene una posición (empezando en 0)### 💡 **¿Cuándo usar listas?**- Para almacenar múltiples elementos relacionados- Cuando necesitas mantener un orden específico- Para iterar sobre colecciones de datos

````python
# 🌍 Creando una lista de paísespaises = ["Colombia", "México", "Argentina", "España", "Perú"]print("🌍 Lista de países:", paises)print("📊 Tipo de dato:", type(paises))print("📏 Cantidad de países:", len(paises))# 📍 Accediendo a elementos por índiceprint("\n📍 Acceso por índice:")print("🥇 Primer país (índice 0):", paises[0])print("🥈 Segundo país (índice 1):", paises[1])print("🏆 Último país (índice -1):", paises[-1])print("🥉 Penúltimo país (índice -2):", paises[-2])
````

### 4.1 ✂️ SlicingEl slicing te permite obtener porciones de una lista. ¡Es como cortar rebanadas de un pastel! 🍰

````python
# ✂️ Ejemplos de slicingprint("📋 Lista completa:", paises)print("🔢 Índices:        ", [0, 1, 2, 3, 4])print("\n✂️ Ejemplos de slicing:")print("🎯 Primeros 3 países [0:3]:", paises[0:3])print("🎯 Desde índice 2 [2:]:", paises[2:])print("🎯 Hasta índice 3 [:3]:", paises[:3])print("🎯 Últimos 2 países [-2:]:", paises[-2:])print("🎯 Todos excepto el último [:-1]:", paises[:-1])# 🦘 Slicing con pasosprint("\n🦘 Slicing con pasos:")print("🎯 Cada 2 elementos [::2]:", paises[::2])print("🔄 Lista invertida [::-1]:", paises[::-1])
````

### 4.2 ➕ Agregar/Eliminar ElementosLas listas son dinámicas. ¡Puedes agregar y quitar elementos cuando quieras! 🔄

## 5. 📖 DiccionariosLos diccionarios almacenan datos en pares clave-valor. ¡Son como un directorio telefónico digital! 📞### 🎯 **Características de los Diccionarios:**- **🔑 Clave-Valor**: Cada elemento tiene una clave única y un valor- **🚀 Rápidos**: Acceso súper eficiente a los datos- **🔄 Mutables**: Puedes modificarlos después de crearlos- **🎭 Flexibles**: Las claves pueden ser strings, números, etc.### 💡 **¿Cuándo usar diccionarios?**- Para representar objetos con propiedades (como una persona con nombre, edad, etc.)- Cuando necesitas acceso rápido por clave- Para mapear relaciones entre datos

````python
# 👤 Creando un diccionario de perfil de usuarioperfil_usuario = {    "nombre": "Ana García",    "edad": 28,    "ciudad": "Madrid",    "profesion": "Desarrolladora",    "lenguajes": ["Python", "JavaScript", "Java"],    "activo": True,    "puntuacion": 95.5}print("👤 Perfil completo:")print(perfil_usuario)print(f"\n📊 Tipo de dato: {type(perfil_usuario)}")print(f"📏 Cantidad de campos: {len(perfil_usuario)}")# 🔍 Accediendo a valores específicosprint("\n🔍 Acceso a valores:")print(f"👋 Nombre: {perfil_usuario['nombre']}")print(f"🎂 Edad: {perfil_usuario['edad']} años")print(f"🏙️ Ciudad: {perfil_usuario['ciudad']}")print(f"💼 Profesión: {perfil_usuario['profesion']}")print(f"🐍 Lenguajes: {perfil_usuario['lenguajes']}")
````

````python
# 🔍 Métodos útiles de diccionariosprint("🔑 Todas las claves:", list(perfil_usuario.keys()))print("📋 Todos los valores:", list(perfil_usuario.values()))print("🔗 Pares clave-valor:", list(perfil_usuario.items()))# 🛡️ Acceso seguro con get()print(f"\n🛡️ Acceso seguro:")print(f"📧 Email: {perfil_usuario.get('email', 'No especificado')}")print(f"📱 Teléfono: {perfil_usuario.get('telefono', 'No disponible')}")# ✅ Verificar si existe una claveprint(f"\n✅ Verificaciones:")print(f"¿Tiene nombre? {'nombre' in perfil_usuario}")print(f"¿Tiene email? {'email' in perfil_usuario}")print(f"¿Está activo? {perfil_usuario.get('activo', False)}")
````

### 5.1 🔄 Agregar/ActualizarLos diccionarios son muy flexibles para modificar datos. ¡Puedes agregar y actualizar fácilmente! ✨

````python
# 🔄 Modificando diccionariosperfil_copia = perfil_usuario.copy()print("📋 Perfil original:")print(f"👤 Nombre: {perfil_copia['nombre']}")print(f"🐍 Lenguajes: {perfil_copia['lenguajes']}")# ✏️ Actualizar valores existentesperfil_copia["nombre"] = "Ana García López"perfil_copia["edad"] = 29# ➕ Agregar nuevos camposperfil_copia["email"] = "ana.garcia@email.com"perfil_copia["telefono"] = "+34 123 456 789"print("\n✨ Perfil actualizado:")print(f"👤 Nombre: {perfil_copia['nombre']}")print(f"🎂 Edad: {perfil_copia['edad']}")print(f"📧 Email: {perfil_copia['email']}")print(f"📱 Teléfono: {perfil_copia['telefono']}")# 🔄 Actualizar múltiples campos con update()nuevos_datos = {    "ciudad": "Barcelona",    "lenguajes": ["Python", "JavaScript", "TypeScript", "Go"],    "experiencia": "5 años"}perfil_copia.update(nuevos_datos)print(f"\n🔄 Después de update():")print(f"🏙️ Ciudad: {perfil_copia['ciudad']}")print(f"🐍 Lenguajes: {perfil_copia['lenguajes']}")print(f"💼 Experiencia: {perfil_copia['experiencia']}")
````

## 6. 🤔 CondicionalesLos condicionales permiten que tu programa tome decisiones. ¡Es como darle un cerebro a tu código! 🧠### 🎯 **Estructura básica:**- **`if`**: Si se cumple la condición- **`elif`**: Si no, pero se cumple otra condición  - **`else`**: Si no se cumple ninguna condición### 🔍 **Operadores de comparación:**- `==` igual a- `!=` diferente de- `>` mayor que- `<` menor que- `>=` mayor o igual- `<=` menor o igual

````python
# 🎂 Ejemplo: Clasificador de edadesedad = 25if edad < 13:    categoria = "👶 Niño"elif edad < 20:    categoria = "🧒 Adolescente"elif edad < 60:    categoria = "👨 Adulto"else:    categoria = "👴 Adulto mayor"print(f"Con {edad} años, eres: {categoria}")# 🌡️ Ejemplo: Clasificador de temperaturatemperatura = 22if temperatura < 0:    estado = "🧊 Congelado"elif temperatura < 15:    estado = "❄️ Frío"elif temperatura < 25:    estado = "😊 Agradable"elif temperatura < 35:    estado = "☀️ Caliente"else:    estado = "🔥 Muy caliente"print(f"Con {temperatura}°C, el clima está: {estado}")# 🔍 Verificar membresía en listaspaises_favoritos = ["España", "Colombia", "México"]pais_buscar = "Colombia"if pais_buscar in paises_favoritos:    print(f"✅ {pais_buscar} está en tu lista de países favoritos")else:    print(f"❌ {pais_buscar} no está en tu lista de países favoritos")
````

## 7. 🔄 Bucles ForLos bucles for te permiten repetir acciones automáticamente. ¡Es como tener un asistente incansable! 🤖### 🎯 **¿Para qué sirven?**- Procesar todos los elementos de una lista- Repetir una acción un número específico de veces- Transformar o analizar datos en masa### 💡 **Funciones útiles:**- `enumerate()`: Obtiene índice y valor- `range()`: Genera secuencias de números- `zip()`: Combina múltiples listas

````python
# 🌍 Iterando sobre listaspaises = ["España", "México", "Colombia", "Argentina"]print("🌍 Recorriendo países:")for pais in paises:    print(f"  🏳️ {pais}")# 🔢 Usando enumerate para obtener índicesprint("\n📊 Con índices:")for indice, pais in enumerate(paises):    print(f"  {indice + 1}. {pais}")# 🎯 Bucle con condicionesprint("\n🔍 Buscando países específicos:")for pais in paises:    if pais == "Colombia":        print(f"  ✅ ¡Encontré {pais}!")    else:        print(f"  ⏭️ {pais} no es Colombia")# 🔢 Usando range() para númerosprint("\n🔢 Contando del 1 al 5:")for numero in range(1, 6):    print(f"  Número: {numero}")# 🎲 Tabla de multiplicarnumero_base = 7print(f"\n🎲 Tabla del {numero_base}:")for i in range(1, 11):    resultado = numero_base * i    print(f"  {numero_base} × {i} = {resultado}")
````

````python
# 📖 Iterando sobre diccionariosestudiante = {    "nombre": "Carlos",    "edad": 22,    "carrera": "Ingeniería",    "promedio": 8.5,    "materias": ["Matemáticas", "Física", "Programación"]}print("👤 Información del estudiante:")for clave, valor in estudiante.items():    if clave == "materias":        print(f"  📚 {clave.title()}: {', '.join(valor)}")    else:        print(f"  📋 {clave.title()}: {valor}")# 🎯 Procesamiento de datos con buclesprint("\n📊 Análisis de calificaciones:")calificaciones = [8.5, 9.2, 7.8, 9.5, 8.0, 9.8, 7.5]suma_total = 0calificaciones_altas = []for calificacion in calificaciones:    suma_total += calificacion    if calificacion >= 9.0:        calificaciones_altas.append(calificacion)promedio = suma_total / len(calificaciones)print(f"  📈 Promedio: {promedio:.2f}")print(f"  🏆 Calificaciones altas (≥9.0): {calificaciones_altas}")print(f"  📊 Total de calificaciones altas: {len(calificaciones_altas)}")
````

### 8.1 🏗️ Funciones BásicasEmpezemos con funciones simples y útiles:

````python
# 👋 Función simple de saludodef saludar(nombre):    """Función que saluda a una persona por su nombre"""    return f"¡Hola {nombre}! 👋 ¡Bienvenido a Python!"# 🧮 Función de sumadef sumar(a, b):    """Suma dos números y devuelve el resultado"""    resultado = a + b    return resultado# 🎂 Función para calcular edaddef calcular_edad_en_dias(edad_anos):    """Convierte años a días aproximados"""    dias = edad_anos * 365    return dias# 🌡️ Función de conversión de temperaturadef celsius_a_fahrenheit(celsius):    """Convierte temperatura de Celsius a Fahrenheit"""    fahrenheit = (celsius * 9/5) + 32    return fahrenheit# 🧪 Probando las funcionesprint("🧪 Probando nuestras funciones:")print(saludar("Ana"))print(f"➕ 15 + 25 = {sumar(15, 25)}")print(f"🎂 25 años son aproximadamente {calcular_edad_en_dias(25):,} días")print(f"🌡️ 25°C = {celsius_a_fahrenheit(25)}°F")
````

### 8.2 🎯 Funciones AvanzadasFunciones más sofisticadas con múltiples parámetros y valores por defecto:

````python
# 📊 Función con múltiples parámetros y valores por defectodef analizar_lista_numeros(numeros, mostrar_detalles=True):    """    Analiza una lista de números y devuelve estadísticas básicas        Args:        numeros (list): Lista de números a analizar        mostrar_detalles (bool): Si mostrar información detallada        Returns:        dict: Diccionario con las estadísticas    """    if not numeros:        return {"error": "Lista vacía"}        estadisticas = {        "total": len(numeros),        "suma": sum(numeros),        "promedio": sum(numeros) / len(numeros),        "maximo": max(numeros),        "minimo": min(numeros)    }        if mostrar_detalles:        print(f"📊 Análisis de {len(numeros)} números:")        print(f"  ➕ Suma total: {estadisticas['suma']}")        print(f"  📈 Promedio: {estadisticas['promedio']:.2f}")        print(f"  🔝 Máximo: {estadisticas['maximo']}")        print(f"  🔻 Mínimo: {estadisticas['minimo']}")        return estadisticas# 🎮 Función para crear perfil de jugadordef crear_perfil_jugador(nombre, nivel=1, puntos=0, activo=True):    """Crea un perfil de jugador con valores por defecto"""    perfil = {        "nombre": nombre,        "nivel": nivel,        "puntos": puntos,        "activo": activo,        "fecha_creacion": "2024-01-15"  # En un caso real, usarías datetime    }        print(f"🎮 Perfil creado para {nombre}:")    print(f"  🏆 Nivel: {nivel}")    print(f"  ⭐ Puntos: {puntos}")    print(f"  ✅ Estado: {'Activo' if activo else 'Inactivo'}")        return perfil# 🧪 Probando las funciones avanzadasprint("🧪 Probando funciones avanzadas:\n")# Análisis de númerosnumeros_prueba = [85, 92, 78, 96, 88, 91, 83]resultado = analizar_lista_numeros(numeros_prueba)print(f"\n📋 Resultado completo: {resultado}")# Creando perfiles de jugadoresprint("\n" + "="*50)jugador1 = crear_perfil_jugador("Alex")print()jugador2 = crear_perfil_jugador("María", nivel=5, puntos=1250)
````

## 9. 📁 MódulosLos módulos extienden las capacidades de Python. ¡Es como tener una caja de herramientas gigante! 🧰### 🎯 **¿Qué son los módulos?**- **📦 Bibliotecas**: Colecciones de funciones pre-escritas- **🚀 Eficiencia**: No reinventes la rueda- **🌍 Comunidad**: Miles de módulos disponibles- **🔧 Especialización**: Herramientas para tareas específicas### 💡 **Módulos populares:**- **`os`**: Interacción con el sistema operativo- **`math`**: Operaciones matemáticas avanzadas- **`random`**: Generación de números aleatorios- **`datetime`**: Manejo de fechas y tiempo

### 9.1 🗂️ Módulo OSEl módulo `os` te permite interactuar con el sistema operativo:

````python
# 🗂️ Trabajando con el módulo OSimport osprint("🗂️ Explorando el sistema con el módulo OS:")# 📍 Directorio actualdirectorio_actual = os.getcwd()print(f"📍 Directorio actual: {directorio_actual}")# 📋 Listar archivos y carpetascontenido = os.listdir()print(f"📋 Contenido del directorio ({len(contenido)} elementos):")for elemento in contenido[:5]:  # Mostramos solo los primeros 5    print(f"  📄 {elemento}")if len(contenido) > 5:    print(f"  ... y {len(contenido) - 5} elementos más")# 🏠 Información del usuariousuario = os.getenv('USER') or os.getenv('USERNAME') or 'Usuario'print(f"👤 Usuario actual: {usuario}")# 🖥️ Sistema operativosistema = os.nameprint(f"🖥️ Sistema operativo: {sistema}")# 📁 Crear directorio (con manejo seguro)nombre_carpeta = "mi_proyecto_python"try:    os.makedirs(nombre_carpeta, exist_ok=True)    print(f"✅ Carpeta '{nombre_carpeta}' creada/verificada")except Exception as e:    print(f"❌ Error al crear carpeta: {e}")# 🔍 Verificar si existeif os.path.exists(nombre_carpeta):    print(f"✅ La carpeta '{nombre_carpeta}' existe")        # 📊 Información adicional    if os.path.isdir(nombre_carpeta):        print(f"📁 Es un directorio")    elif os.path.isfile(nombre_carpeta):        print(f"📄 Es un archivo")
````
